```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
recall = [0.3395, 0.3258, 0.494, 0.4936, 0.5297, 0.5034, 0.5422, 0.5344]
recall_std = [0.0747, 0.0598, 0.0819, 0.082, 0.0837, 0.0834, 0.0845, 0.082]

f1 = [0.3362, 0.3195, 0.1928, 0.1929, 0.2036, 0.196, 0.2102, 0.2034]
f1_std = [0.0523, 0.0436, 0.0427, 0.0427, 0.0404, 0.0424, 0.0429, 0.0401]

sbert = [0.5388, 0.4661, 0.4472, 0.4471, 0.4806, 0.4612, 0.4841, 0.484]
sbert_std = [0.0782, 0.0719, 0.0784, 0.0784, 0.0762, 0.0721, 0.0753, 0.0726]

x_pos = np.arange(len(recall))
models = ["LeadN", "RandomN", "T5-Unmodifed", "T5-F.T.", "T5-Aug-F.T.", "T5-Wiki-Transfer", "T5-Wiki-FT", "T5-Wiki-Aug-FT"]

output_path = "final_results_overall/"
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, sbert, yerr=sbert_std, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('SBERT Cosine-Sim.')
ax.set_xticks(x_pos)
ax.set_xticklabels(models, rotation=45)
ax.set_title('SBERT Embedding Cosine-Similarity (stddev) vs. Model')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig(output_path + "sbert_all.png")
plt.show()
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, f1, yerr=f1_std, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Rouge1 F1-Score')
ax.set_xticks(x_pos)
ax.set_xticklabels(models, rotation=45)
ax.set_title('Rouge1 F1 Score (stddev) vs. Model')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig(output_path + "r1_f1_all.png")
plt.show()
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, recall, yerr=recall_std, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Rouge1 Recall Score')
ax.set_xticks(x_pos)
ax.set_xticklabels(models, rotation=45)
ax.set_title('Rouge1 Recall Score (stddev) vs. Model')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig(output_path + "r1_r_all.png")
plt.show()
```
